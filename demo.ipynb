```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio
import warnings
warnings.filterwarnings('ignore')
from google.colab import drive
drive.mount('/content/gdrive')
# load and evaluate a saved model
from numpy import loadtxt
from keras.models import load_model

# load model
model = load_model('/content/gdrive/Shareddrives/DATA255/model_comb_noise_pitch_adam.h5')
def extract_mfcc(filename):
        
    y, sr = librosa.load(filename, duration=3, offset=0.5)
    
    mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40).T, axis=0)
    return mfcc
demo_path='/content/gdrive/Shareddrives/DATA255/demo10.wav'
Audio(demo_path)
```

```python
X_demo = extract_mfcc(demo_path)
#print(X_demo.shape)
X_demo = np.reshape(X_demo, (1,40,1))
y_demo=model.predict(X_demo)
#print(y_demo)
#print(np.max(y_demo))
emotions=['fear', 'sad', 'angry', 'disgust', 'surprised', 'happy', 'calm']
dframe = pd.DataFrame({'Emotions':emotions})  
dframe['Percentage']=np.reshape(y_demo,-1)
dframe['Percentage']=dframe['Percentage'].mul(100).round(2)
dframe.sort_values(by=['Percentage'],ascending=False)
```
